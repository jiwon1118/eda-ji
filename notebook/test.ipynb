```python
from president_speech.db.parquet_interpreter import get_parquet_full_path
import pandas as pd
```

```python
keyword = "자유"
ascen = False
```

```python
data_path = get_parquet_full_path()
```

```python
df = pd.read_parquet(data_path)  # 데이터 프레임으로 로드
```

```python
df
```

```python
fdf = df[df['speech_text'].str.contains(keyword, case=False)] # 키워드가 들어간 연설 필터링
```

```python
fdf
```

```python
gdf = fdf.groupby("president").size().reset_index(name="count")  # 대통령별 그룹화 하고 개수 세기
```

```python
gdf
```

```python
sdf = gdf.sort_values(by='count', ascending=ascen).reset_index(drop=True)  # 정렬
```

```python
sdf
```

```python
presidents_speeches = {
    "박정희": 513,
    "이승만": 438,
    "노태우": 399,
    "김대중": 305,
    "문재인": 275,
    "김영삼": 274,
    "이명박": 262,
    "전두환": 242,
    "노무현": 230,
    "박근혜": 111,
    "최규하": 14,
    "윤보선": 1
}
```

```python
for p_name, s_count in presidents_speeches.items():
    print(f"{p_name} -> {s_count}")
```

```python
for p_name, s_count in presidents_speeches.items():
    president_row = sdf[sdf["president"] == p_name]
    print(president_row.iloc[0]["count"] == s_count) 
```

```python
for p_name, s_count in presidents_speeches.items():
    president_row = sdf[sdf["president"] == p_name]
    is_ok = president_row.iloc[0]["count"] == s_count
    print(f"{p_name} = {is_ok}")
```
