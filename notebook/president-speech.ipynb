### QUIZ

speech_text 에 "올림픽"이 포함된 ROW를 필터링 후 대통령 카운트 내림차순으로 sorting

```python
f_df = df[df['speech_text'].str.contains("올림픽", regex=False)]
```

```python
f_df
```

```python
rdf = f_df.groupby("president").size().reset_index(name='count')
```

```python
rdf
```

```python
rdf.sort_values(by='count',ascending=False)
```

```python
sdf = rdf.sort_values(by='count', ascending=False).reset_index(drop=True)
```

```python
print(sdf.to_string(index=False))
```
